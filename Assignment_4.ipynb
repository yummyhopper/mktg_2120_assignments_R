Assuming that the interest rate = 0 and that the probability of churn stays constant over time, the CLV formula can be written as: CLV = r x m (1/Churn Rate). Because, 1 - r = Churn Rate, we can assume 1 - Churn = r. Therefore, for customer #1, we used the same coefficient estimates and churn rate we found from Assignment 2’s logistic regression model. This allowed us to create the logarithmic function below, which we subtracted from 1 to get r = 0.744. We then plugged that into the original formula. The margin is $55 because we gain $60 in revenue per month from the customer, but lose $5 in maintenance cost monthly. This was multiplied by (r1-r) to get CLV = $160.21.

```r
# retention rate person 1

retention_person_1 = 1-(1 / (1 + exp(-(-0.711606+(-.035382*20)-0.827540+0.831927-1.002658+(60*0.014895)+0.454616))))
retention_person_1
```

```r
# clv for person 1

clv_person_1 = 55*(retention_person_1/(1+0-retention_person_1))
clv_person_1
```

```r
# retention rate person 2

retention_person_2 = 1-(1 / (1 + exp(-(-0.711606+(-.035382*20)-0.827540+0.831927-1.002658+(60*0.014895)+0))))
retention_person_2
```

### 2d.

The company is considering no longer investing in paid search advertising. Test whether there is a statistically significant difference in conversion rate between the PaidSearch and OrganicSearch campaigns. Report which test you are using, what significance cutoff you are using, what the null and alternative hypotheses are, the results, and your conclusions.

---

```r
paid_search <- data %>% filter(campaign_type == "PaidSearch")
organic_search <- data %>% filter(campaign_type == "OrganicSearch")

# Perform two-sample proportion test
test_result <- prop.test(x = c(sum(paid_search$converted == "Yes"), sum(organic_search$converted == "Yes")),
                         n = c(nrow(paid_search), nrow(organic_search)))
test_result
```

* To test for a statistically significant difference in conversion rate between `PaidSearch` and `OrganicSearch` campaigns, we can use a two-sample proportion test.

* Null hypothesis: There is no difference in conversion rate between `PaidSearch` and `OrganicSearch` campaigns.

* Alternative hypothesis: There is a difference in conversion rate between `PaidSearch` and `OrganicSearch` campaigns.

* Significance cutoff: 0.05

* P-value is smaller than 0.05, so we reject the null. There is a statistically significant difference in conversion rate between `PaidSearch` and `OrganicSearch`.
